```python
import requests
import datetime as dt
import pandas as pd
from pandas.io.json import json_normalize
import json
from flatten_json import flatten
```

```python
def getCoinList():
    req = requests.get('https://www.cryptocompare.com/api/data/coinlist/').json()
    info = req['Data']
    coinList = pd.DataFrame(info)
    coinList = coinList.transpose()
    coinList.to_csv('coinList.csv')
    return coinList
```

```python
def getSocialCoinDetails(coins):
    details = []
    for coinID in coins:
        try:
            req = requests.get('https://www.cryptocompare.com/api/data/socialstats/?id='+str(coinID)).json()
            details.append(req)
        except:
            pass
    return pd.DataFrame(details)
```

```python
coinList_df = getCoinList()
myList = ["BTC", "ETH", "XRP","BCH","LTC",
          "ADA","EOS","XLM","NEO","IOT",
          "VEN","DASH","BCD","XEM","TRX"]
```

```python
topCoinList_df = coinList_df.loc[coinList_df['Name'].isin(myList)]
topCoinList_df.head(15)
```

```python
ids = list(topCoinList_df['Id'])
socialInfo = getSocialCoinDetails(ids)
x = [flatten(Twitter) for Twitter in socialInfo['Data']]
social_df = json_normalize(x)
```

```python
social_condensed_df = social_df[['General_CoinName','General_Name','General_Points','General_Type',
                                 'Reddit_Points','Reddit_active_users','Reddit_comments_per_day',
                                 'Reddit_comments_per_hour','Reddit_community_creation',
                                 'Reddit_name','Reddit_posts_per_day','Reddit_posts_per_hour',
                                 'Reddit_subscribers','Twitter_Points','Twitter_account_creation',
                                 'Twitter_favourites','Twitter_followers','Twitter_following',
                                 'Twitter_lists','Twitter_name','Twitter_statuses','Facebook_Points',
                                 'Facebook_is_closed','Facebook_likes','Facebook_name',
                                 'Facebook_talking_about','CryptoCompare_Comments','CryptoCompare_PageViews',
                                 'CryptoCompare_PageViewsSplit_Analysis','CryptoCompare_PageViewsSplit_Charts',
                                 'CryptoCompare_PageViewsSplit_Forum','CryptoCompare_PageViewsSplit_Influence',
                                 'CryptoCompare_PageViewsSplit_Markets','CryptoCompare_PageViewsSplit_Orderbook',
                                 'CryptoCompare_PageViewsSplit_Overview','CryptoCompare_PageViewsSplit_Trades',
                                 'CryptoCompare_Points','CryptoCompare_Posts','CodeRepository_Points']]
```

```python
social_condensed_df['Date'] = dt.datetime.today().strftime("%m/%d/%Y")
```
