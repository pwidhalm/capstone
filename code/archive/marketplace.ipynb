```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import talib as ta
import numpy as np
```

```python
btc_df = pd.read_csv("../data/btc.csv", index_col=0)
btc_df.head()
```

```python
ltc_df = pd.read_csv("../data/ltc.csv", index_col=0)
ltc_df.head()
```

```python
vtc_df = pd.read_csv("../data/vtc.csv", index_col=0)
vtc_df.head()
```

```python
btc_df = btc_df.drop(['txVolume(USD)', 'txCount', 'marketcap(USD)', 'exchangeVolume(USD)', 'generatedCoins', 'fees'], axis=1)
btc_df.columns = ['btc']
ltc_df = ltc_df.drop(['txVolume(USD)', 'txCount', 'marketcap(USD)', 'exchangeVolume(USD)', 'generatedCoins', 'fees'], axis=1)
ltc_df.columns = ['ltc']
vtc_df = vtc_df.drop(['txVolume(USD)', 'txCount', 'marketcap(USD)', 'exchangeVolume(USD)', 'generatedCoins', 'fees'], axis=1)
vtc_df.columns = ['vtc']
```

```python
# use VTC index as date range. 
df = pd.concat([btc_df, ltc_df, vtc_df], axis=1, join_axes=[vtc_df.index])
df.head()
```

```python
df_pct = df.pct_change()
df_pct.dropna(inplace=True)
df_pct.head()
```

```python
df_pct.to_csv('./market-returns.csv', index=False)
```
