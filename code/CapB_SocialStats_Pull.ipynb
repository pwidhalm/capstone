```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('fivethirtyeight')
# Pretty print the JSON
import uuid
from IPython.display import display_javascript, display_html, display
import json

class RenderJSON(object):
    def __init__(self, json_data):
        if isinstance(json_data, dict):
            self.json_str = json.dumps(json_data)
        else:
            self.json_str = json_data
        self.uuid = str(uuid.uuid4())

    def _ipython_display_(self):
        display_html('<div id="{}" style="height: 600px; width:100%;"></div>'.format(self.uuid), raw=True)
        display_javascript("""
        require(["https://rawgit.com/caldwell/renderjson/master/renderjson.js"], function() {
        document.getElementById('%s').appendChild(renderjson(%s))
        });
        """ % (self.uuid, self.json_str), raw=True)
```

```python
def getCoinList():
    req = requests.get('https://www.cryptocompare.com/api/data/coinlist/').json()
    info = req['Data']
    coinList = pd.DataFrame(info)
    coinList = coinList.transpose()
    coinList.to_csv('coinList.csv')
    return coinList
```

```python
df = getCoinList()
df.head(25)
```

```python
def getSocialCoinDetails(coins):
    details = []
    for coinID in coins:
        try:
            req = requests.get('https://www.cryptocompare.com/api/data/socialstats/?id='+str(coinID)).json()['Data']['General']
            details.append(req)
        except:
            pass
    return pd.DataFrame(details)
```
